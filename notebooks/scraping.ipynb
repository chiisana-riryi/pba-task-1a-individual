```python
from google_play_scraper import reviews, Sort
from pprint import pprint
import csv
```

```python
# constants
app_id : str = "com.valvesoftware.android.steam.community"
scrape_review_count : int = 2000
```

```python
reviews_csv_file_dir : str = "../data/reviews-%-stars.csv"
```

```python
def scrape_reviews(
        count : int = 100,
        score : int = None,
        sort  : Sort = Sort.MOST_RELEVANT, 
        lang  : str = "en",
    ) -> Tuple[List[dict], _ContinuationToken]:

    return reviews(
        app_id = app_id,
        lang   = lang,
        sort   = sort,
        count  = count,
        filter_score_with  = score,
    )
```

```python
# test
data, _ = scrape_reviews(count=1)

pprint(data)
```

```python
# scrape and store to csv

for i in range (0, 5):
    stars = i + 1
    data, _ = scrape_reviews(count=scrape_review_count, score=stars)

    reviews_csv_file_dir_star = reviews_csv_file_dir.replace("%", str(stars))
    reviews_csv_file = open(reviews_csv_file_dir_star, "w")

    if (len(data) > 0):
        writer = csv.writer(reviews_csv_file)

        header_keys = data[0].keys()
        writer.writerow(header_keys)
        
        for review in data:
            writer.writerow(review.values())
```

```python
# close files

reviews_csv_file.close()
```
